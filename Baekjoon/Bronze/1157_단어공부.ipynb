```python
str = input().upper()


counts = {}

for char in str:
    if char in counts:
        counts[char] += 1
    else:
        counts[char] = 1


max_val = max(counts.values())

result = [k for k, v in counts.items() if v == max_val]

if len(result) > 1:
    print("?")
else:
    print(result[0])
```
