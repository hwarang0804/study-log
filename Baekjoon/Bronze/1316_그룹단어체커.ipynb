```python
N = int(input())


count = 0


for i in range(N):
    word = input()
    memory = []
    for i in range(0,len(word)):
        if word[i] not in memory: 
            memory.append(word[i])
        elif word[i] in memory:
            if word[i] != word[i - 1]:
                break
        if i == len(word) - 1:
            count += 1
print(count)

            



#앞에 글자를 기억하면서 가야함
#기억해 가는데, 다음 글자가 새로운 글자가
#아니면서 이전에 기억된 적이 있으면 제외
#아니라면 연속된 글자로서 카운트

```
