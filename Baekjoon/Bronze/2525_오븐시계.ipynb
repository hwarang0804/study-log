```python
A, B = map(int, input().split())
C = int(input())

hour = C // 60
minute = C % 60

if B + minute >= 60:
    A = A + 1
    B = B - 60
    if A + hour >= 24:
        A = A - 24
        print(A + hour, B + minute)
    else:
        print(A + hour, B + minute)
elif A + hour >= 24:
    A = A - 24
    print(A + hour, B + minute)
else:
    print(A + hour, B + minute)
```
