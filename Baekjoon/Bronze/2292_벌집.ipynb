```python
n = int(input())

# 1층부터 시작
honeycomb_max = 1  # 해당 층에서 가장 큰 숫자
count = 1          # 지나가는 방의 최소 개수 (층수)

# 입력받은 n이 현재 층의 최대 숫자보다 클 때만 반복
while n > honeycomb_max:
    # 다음 층의 최대 숫자는 (6 * 현재 층수)만큼 늘어남
    honeycomb_max += 6 * count
    # 층수(지나가는 방의 개수) 증가
    count += 1

print(count)
```
