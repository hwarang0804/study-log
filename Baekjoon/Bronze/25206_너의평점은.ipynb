```python
import sys
input = sys.stdin.readline

sum_gakuten = 0
sum = 0
ranks = {
    "A+" : 4.5,
    "A0" : 4.0,
    "B+" : 3.5,
    "B0" : 3.0,
    "C+" : 2.5,
    "C0" : 2.0,
    "D+" : 1.5,
    "D0" : 1.0,
    "F" : 0.0,
}

for _ in range(20):
    entire = input().split()
    gakuten = float(entire[1])
    rank = entire[2]
    if rank in ranks:
        rank_number = ranks[rank]
        sum = sum + (gakuten * rank_number)
        sum_gakuten = sum_gakuten + gakuten
result = sum / sum_gakuten
print(result)

    
    
    
```
