```python
K, Q, L, B, KN, P = map(int, input().split())
correct_answer = [1, 1, 2, 2, 2, 8]
lists = [K, Q, L, B, KN, P]
box = []
for i in range(len(correct_answer)):
        if correct_answer[i] == lists[i]:
            box.append(0)
        elif correct_answer[i] > lists[i]:
            box.append(correct_answer[i] - lists[i])
        elif correct_answer[i] < lists[i]:
            box.append(correct_answer[i] - lists[i])
            
print(*box)
```
