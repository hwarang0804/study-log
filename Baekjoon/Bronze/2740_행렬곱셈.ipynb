```python
N,M = map(int, input().split())

matrixA = []
matrixB = []

for _ in range(N):
    lineA = list(map(int, input().split()))
    matrixA.append(lineA)

M,K = map(int, input().split())

for _ in range(M):
    lineB = list(map(int, input().split()))
    matrixB.append(lineB)

for i in range(N):
    line = []
    for j in range(K):
        sum = 0
        for x in range(M):    
            sum = sum + matrixA[i][x] * matrixB[x][j]    
        line.append(sum)
    print(*line)
```
