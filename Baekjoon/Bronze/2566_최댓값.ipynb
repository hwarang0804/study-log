```python
max_r, max_c = 9, 9

matrix = []
memory = []
maxs = []

for _ in range(max_r):
    line = list(map(int, input().split()))
    matrix.append(line)

for row in range(max_r):
    for column in range(max_c):
        if max(matrix[row]) == matrix[row][column]:
            memory.append([row, column])
    maxs.append(max(matrix[row]))
print(max(maxs))
for i in range(len(memory)):
    if max(maxs) == matrix[memory[i][0]][memory[i][1]]:
        print(memory[i][0] + 1,memory[i][1] + 1)
        break

```

```python
max_val = 0
max_raw = 0
max_col = 0

for row in range(9):
    current_row = list(map(int, input().split()))
    for column in range(9):
        if current_row[column] > max_val:
            max_val = current_row[column]
            max_raw = row
            max_col = column

print(max_val)
print(max_raw + 1, max_col + 1)
```
