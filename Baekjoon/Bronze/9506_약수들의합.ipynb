```python
while True:
    N = int(input())
    
    factors = []

    if N == -1:
        break

    for i in range(N):
        if N == (i + 1):
            break
        if N % (i + 1) == 0:
            factors.append(i + 1)
    
    if sum(factors) == N:
        total = 0
        print(f"{N} = ", end="")
        for j in range(N):
            if N == (j + 1):
                break
            if N % (j + 1) == 0:
                total += j + 1
                if total == sum(factors):
                    print(f"{j + 1}")
                else:
                    print(f"{j + 1} + ", end="")
    else:
        print(f"{N} is NOT perfect.")


    
```
