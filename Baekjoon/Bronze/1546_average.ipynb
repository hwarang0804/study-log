```python
N = int(input())

scores = list((map(int, input().split())))

max_score = 0

for i in range(N):
    if max_score < scores[i]:
        max_score = scores[i]

scores_sum = 0
for j in range(N):
    scores[j] = scores[j] / max_score * 100
    scores_sum = scores_sum + scores[j]
average = float(scores_sum) / N
print(average)
```
