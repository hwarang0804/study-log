```python
N, M = map(int, input().split())


#sum_matrix = []
matrixA = []
matrixB = []

for _ in range(N):
    lineA = list(map(int, input().split()))
    matrixA.append(lineA)

for _ in range(N):
    lineB = list(map(int, input().split()))
    matrixB.append(lineB)

for i in range(N):
    line = []
    for j in range(M):
        line.append(matrixA[i][j] + matrixB[i][j])
    print(*line)
    #sum_matrix.append(line)
#print(sum_matrix)
```
