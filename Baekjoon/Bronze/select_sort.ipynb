```python
li = [8 ,7, 4, 5, 6, 3, 2, 1]
n = len(li)
i = 0

while i < n - 1:
    j = i + 1
    indexOfSmall = i
    while j < n:
        if li[indexOfSmall] > li[j]:
            indexOfSmall = j
        j += 1
    li [i], li[indexOfSmall] = li[indexOfSmall], li[i]
    i += 1
li
```
