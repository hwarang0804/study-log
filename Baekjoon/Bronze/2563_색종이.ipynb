```python
paper = []

for _ in range(100):
    line = []
    for _ in range(100):
        line.append(0)
    paper.append(line)

N = int(input())

for _ in range(N):
    x, y = map(int, input().split())
    originX = x
    originY = y
    while True:
        x = originX
        while True:
            paper[y - 1][x - 1] = 1
            x += 1
            if originX + 10 <= x:
                break
        y += 1
        if originY + 10 <= y:
            break

total = 0
for i in range(100):
    total = total + sum(paper[i])
print(total)
```

```python
paper = []

for _ in range(100):
    line = []
    for _ in range(100):
        line.append(0)
    paper.append(line)

N = int(input())

for _ in range(N):
    x, y = map(int, input().split())
    for row in range(y, y + 10):
        for col in range(x, x + 10):
            paper[row - 1][col - 1] = 1

total = 0
for i in range(100):
    total = total + sum(paper[i])
print(total)
```
