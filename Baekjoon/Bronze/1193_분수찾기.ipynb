```python
# 합이 2인건 1개
# 합이 3인건 2개
# 합이 4인건 3개
# 합이 5인건 4개
# 합이 6인건 5개


# 만약 내가 14번째 분수를 찾고 싶다고 하면
# 1 2 3 4 5 
# 합이 6인것에서 걸리고,
# 합이 짝수이면 아래에서 위로(분모가 커지고, 분자가 작아짐),
# 합이 홀수이면 위에서 아래로(분모가 작아지고, 분자가 커짐),

# 판단 해야하는 것
# 어느 라인에 있는지
# 해당 라인이 홀수인지 짝수인지
# 해당 라인의 몇번째인지
```

```python
N = int(input())

max_number = 1
count_line = 1

while N > max_number:
    count_line += 1
    max_number += count_line

A = max_number - N

if count_line % 2 != 0:
    print(f"{count_line + 1 - (count_line - A)}/{count_line - A}")
else:
    print(f"{count_line - A}/{count_line + 1 - (count_line - A)}")
```
