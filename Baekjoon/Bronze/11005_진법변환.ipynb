```python
import sys

# 1. 입력 받기 (N: 10진법 수, B: 변환할 진법)
N, B = map(int, sys.stdin.readline().split())

# 2. 나머지가 10 이상일 경우 사용할 딕셔너리 생성
# 10: 'A', 11: 'B', ..., 35: 'Z'
alphabet_dict = {i + 10: chr(ord('A') + i) for i in range(26)}

result = []

# 3. N % B와 N // B를 반복하며 나머지 채우기
while N > 0:
    remainder = N % B
    
    # 나머지가 10 이상이면 딕셔너리에서 알파벳을 가져옴
    if remainder >= 10:
        result.append(alphabet_dict[remainder])
    else:
        result.append(str(remainder))
    
    # 몫을 업데이트
    N //= B

# 4. 나머지를 발생 순서대로 저장했으므로 역순으로 뒤집어서 출력
print("".join(result[::-1]))
```
