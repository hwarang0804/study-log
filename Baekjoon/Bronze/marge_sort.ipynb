```python
def mergeSort(x):
    
    if len(x) <= 1:
        return x

    n = len(x)
    nDivideLeft = x[ :n//2]
    nDivideRight = x[n//2: ]

    leftSorted = mergeSort(nDivideLeft)
    rightSorted = mergeSort(nDivideRight)

    result = []

    i = 0
    j = 0

    while i < len(leftSorted) and j < len(rightSorted):
        if leftSorted[i] < rightSorted[j]:
            result.append(leftSorted[i])
            i += 1
        elif leftSorted[i] > rightSorted[j]:
            result.append(rightSorted[j])
            j += 1
        else:
            result.append(leftSorted[i])
            result.append(rightSorted[j])
            i += 1
            j += 1
    if i == len(leftSorted) or j == len(rightSorted):
        if i == len(leftSorted):
            for val in rightSorted[j :]:
                result.append(val)
        elif j == len(rightSorted):
            for val in leftSorted[i :]:
                result.append(val)


    return result

li = [5, 8, 6, 7, 1, 3, 2, 4,]

sortedLi = mergeSort(li)
print(sortedLi)


```

```python
def merge_sort(arr):
    # 1. Base Case (종료 조건)
    if len(arr) <= 1:
        return arr

    # 2. Divide (분할) - 변수명을 간결하게
    mid = len(arr) // 2
    left_half = arr[:mid]
    right_half = arr[mid:]

    # 3. Conquer (재귀 호출)
    left_sorted = merge_sort(left_half)
    right_sorted = merge_sort(right_half)

    # 4. Merge (병합)
    result = []
    i = 0
    j = 0

    while i < len(left_sorted) and j < len(right_sorted):
        # 등호(<=)를 넣어서 3가지 경우(if-elif-else)를 2가지(if-else)로 줄임
        if left_sorted[i] <= right_sorted[j]:
            result.append(left_sorted[i])
            i += 1
        else:
            result.append(right_sorted[j])
            j += 1
    
    # [핵심 개선 포인트]
    # 남은 요소 처리: 복잡한 if문 없이 슬라이싱으로 한 방에 처리
    # 한쪽이 이미 비어있다면, += []가 되어 아무 일도 일어나지 않으므로 안전함
    result += left_sorted[i:]
    result += right_sorted[j:]

    return result

# 테스트
li = [5, 8, 6, 7, 1, 3, 2, 4]
sorted_li = merge_sort(li)
print(sorted_li)
```
