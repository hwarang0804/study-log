```python
N, B = input().split()
B = int(B) # B는 숫자로 변환해야 계산이 가능합니다.

# A=10부터 시작하는 일반적인 진법 변환용 딕셔너리
alphabet_dict = {char: i + 10 for i, char in enumerate("ABCDEFGHIJKLMNOPQRSTUVWXYZ")}

total = 0
# N을 뒤집어서 0승부터 차례대로 계산
N = N[::-1] 

for i in range(len(N)):
    # 1. 값 추출 (알파벳이면 딕셔너리에서, 숫자면 그대로)
    if N[i] in alphabet_dict:
        value = alphabet_dict[N[i]]
    else:
        value = int(N[i])
    
    # 2. 진법 계산: 값 * (진법의 i제곱)
    total += value * (B ** i)

print(total)
```
