```python
N, M = map(int, input().split())

boxs = [0] * N

for a in range(N):
    boxs[a] = a + 1

for b in range(M):
    select = []
    i, j = map(int, input().split())
    select = [0] * (j - i + 1)
    repeat_number = len(select) // 2
    for c in range(repeat_number):
        boxs[i - 1], boxs[j - 1] = boxs[j - 1], boxs[i - 1]
        i += 1
        j -= 1
print(*boxs)
```
