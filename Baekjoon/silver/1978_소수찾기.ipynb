```python
N = int(input()) 
numbers = list(map(int, input().split()))

count = 0

for i in range(N):
    for j in range(numbers[i]):
        if numbers[i] == (2 + j):
            break
        elif numbers[i] % (2 + j) == 0:
            break
    if numbers[i] == (2 + j):
        count += 1 
print(count)
```
